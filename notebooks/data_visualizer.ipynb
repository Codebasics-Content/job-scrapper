```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from pathlib import Path

# Set style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)
```

```python
# Get all CSV files
csv_dir = Path('data/csv')
csv_files = sorted(csv_dir.glob('*.csv'))
print(f"Found {len(csv_files)} CSV files")
for f in csv_files:
    print(f"  - {f.name}")
```

```python
# Function to create visualization for each CSV
def visualize_skills(csv_path, top_n=20):
    # Extract job title from filename
    job_title = csv_path.stem.split('_2025')[0].replace('_', ' ')
    
    # Read CSV
    df = pd.read_csv(csv_path)
    
    # Take top N skills
    df_top = df.head(top_n)
    
    # Create figure with 2 subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))
    
    # Bar chart - Count
    sns.barplot(data=df_top, y='Skill', x='Count', ax=ax1, palette='viridis')
    ax1.set_title(f'{job_title} - Top {top_n} Skills by Count', fontsize=14, fontweight='bold')
    ax1.set_xlabel('Count', fontsize=12)
    ax1.set_ylabel('Skill', fontsize=12)
    
    # Bar chart - Percentage
    # Remove % sign and convert to float
    df_top['Percentage_val'] = df_top['Percentage'].str.rstrip('%').astype(float)
    sns.barplot(data=df_top, y='Skill', x='Percentage_val', ax=ax2, palette='mako')
    ax2.set_title(f'{job_title} - Top {top_n} Skills by Percentage', fontsize=14, fontweight='bold')
    ax2.set_xlabel('Percentage (%)', fontsize=12)
    ax2.set_ylabel('')
    
    plt.tight_layout()
    plt.show()
    
    # Print summary stats
    print(f"\n{job_title} Summary:")
    print(f"  Total skills: {len(df)}")
    print(f"  Top skill: {df.iloc[0]['Skill']} ({df.iloc[0]['Percentage']})")
    print(f"  Skills above 50%: {len(df[df['Percentage'].str.rstrip('%').astype(float) > 50])}")
    print("-" * 80)
```

## Comparison: Top Skills Across All Roles

```python
# Aggregate top skills across all roles
all_data = []
for csv_file in csv_files:
    job_title = csv_file.stem.split('_2025')[0].replace('_', ' ')
    df = pd.read_csv(csv_file)
    df['Job_Title'] = job_title
    all_data.append(df.head(10))  # Top 10 from each

combined_df = pd.concat(all_data, ignore_index=True)

# Get most common skills
skill_frequency = combined_df['Skill'].value_counts().head(15)

plt.figure(figsize=(12, 8))
sns.barplot(x=skill_frequency.values, y=skill_frequency.index, palette='rocket')
plt.title('Most Frequent Skills Across All Job Roles (Top 15)', fontsize=16, fontweight='bold')
plt.xlabel('Frequency (appears in top 10 of N roles)', fontsize=12)
plt.ylabel('Skill', fontsize=12)
plt.tight_layout()
plt.show()

print("\nMost Common Skills:")
for skill, count in skill_frequency.items():
    print(f"  {skill}: appears in top 10 of {count} roles")
```
